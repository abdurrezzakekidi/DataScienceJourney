```python
from pdf2image import convert_from_path
import matplotlib.pyplot as plt
from PIL import Image

pdf_path1 = r"C:\Users\Abdurezzak Ekidi\Downloads\harita1 Uygulamas.pdf"
pdf_path2 = r"C:\Users\Abdurezzak Ekidi\Downloads\harita2 Uygulamas.pdf"

poppler_path = r"C:\Users\Abdurezzak Ekidi\Downloads\Release-24.07.0-0\poppler-24.07.0\Library\bin"

# PDF'ten resimlere dntrme
pages1 = convert_from_path(pdf_path1, 300, poppler_path=poppler_path)
pages2 = convert_from_path(pdf_path2, 300, poppler_path=poppler_path)

harita1 = pages1[0]
harita2 = pages2[0]

# Haritalar yan yana karlatrmal olarak grntleme
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].imshow(harita1)
axs[0].set_title("Harita 1")
axs[0].axis('off')

axs[1].imshow(harita2)
axs[1].set_title("Harita 2")
axs[1].axis('off')
plt.show()
```

```python
import numpy as np
# Haritalar numpy array'lerine dntr
harita1_array = np.array(harita1)
harita2_array = np.array(harita2)

# ki harita arasndaki fark hesapla
fark = np.abs(harita1_array.astype('int16') - harita2_array.astype('int16'))

# Farkllklar grselletir
plt.imshow(fark, cmap='gray')  # Burada renk haritas deitirilebilir
plt.title("ki Harita Arasndaki Fark")
plt.axis('off')
plt.show()


```

```python
plt.hist(harita1_array.ravel(), bins=256, color='red', alpha=0.5, label='Harita 1')

plt.hist(harita2_array.ravel(), bins=256, color='blue', alpha=0.5, label='Harita 2')

plt.title('Histogram Karlatrmas')
plt.xlabel('Piksel Deeri')
plt.ylabel('Frekans')
plt.legend()
plt.show()

```

```python

blended = Image.blend(harita1, harita2, alpha=0.5) 

# Harmanlanm haritay gster
plt.imshow(blended)
plt.title("Harmanlanm Harita")
plt.axis('off')
plt.show()

```
